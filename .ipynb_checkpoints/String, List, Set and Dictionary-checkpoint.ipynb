```python
#########Python String###########


String1 = "GeeksForGeeks"
print("Initial String: ")
print("\n")
print(String1)
 
# Printing 3rd to 12th character
print("\nSlicing characters from 3-12: ")
print(String1[3:12])
 
# Printing characters between
# 3rd and 2nd last character
print("Slicing characters between " + "3rd and 2nd last character: ")
print("\n")
print(String1[3:-1])
print("\n")

String1 = "Hello, I'm a Geek"
print("Initial String: ")
print(String1)
print(type(String1))
 
# Updating a character
# of the String
String1[2] = ("p")
print("\nUpdating character at 2nd Index: ")
print(String1)
```

```python
#########Python List###########

List = []
for i in range(1, 5):
    List.append(i)
    
print(List)
List.insert(3, 12)
print(List)

List = [['Geeks', 'For'], ['Geeks','sa', 'cd']]
print(List[0][1])
print(List[1][2])


old_string = "aba"
print(old_string)
string_list = list(old_string)
string_list[2] = "c"


new_string = "".join(string_list)

print(string_list)
print(type(string_list))

a=str(string_list)
print(a)
print(type(a))

for i in a:
    print(i)
    
print(new_string)
```

```python
#########Python Set###########

set1 = set("GeeksForGeeks")
print("\nSet with the use of String: ")
print(set1)

"""
    Python'da Set, yinelenebilir, değiştirilebilir ve yinelenen öğeleri olmayan
    sırasız bir veri türü koleksiyonudur. Bir kümedeki öğelerin sırası, çeşitli
    öğelerden oluşabilmesine rağmen tanımsızdır.

    Liste yerine küme kullanmanın en büyük avantajı,
    kümede belirli bir öğenin bulunup bulunmadığını kontrol
    etmek için oldukça optimize edilmiş bir yönteme sahip olmasıdır.
"""

String = 'GeeksForGeeks'
set1 = set(String)
print(set1)
```
